Load the file “zoo.data” and look at the info and the first five rows. The first column denotes the animal name and the last one specifies a high-level class for the corresponding animal.

```python
import pandas as pd

# Column names as per zoo dataset documentation
columns = [
    "animal_name","hair","feathers","eggs","milk","airborne","aquatic",
    "predator","toothed","backbone","breathes","venomous","fins",
    "legs","tail","domestic","catsize","class_type"
]

# Load dataset
df = pd.read_csv("zoo.data", header=None, names=columns)

# Dataset info
print(df.info())

# First five rows
print(df.head())
```

2. Find out the unique number of high-level cclasses

```python
# Unique class labels
unique_classes = df['class_type'].unique()

print("Unique Classes:", unique_classes)
print("Number of High-Level Classes:", len(unique_classes))
```

3. Use the 16-intermediate feature and perform an agglomerative clustering.
[ Hint: Refer to the agglomerative clustering module in scikit learn and set the
number of clusters appropriately]


```python
from sklearn.cluster import AgglomerativeClustering

# Select 16 intermediate features
X = df.iloc[:, 1:-1]   # excludes animal_name & class_type

# Agglomerative Clustering
agglo = AgglomerativeClustering(
    n_clusters=7,
    linkage='ward'
)

# Fit and predict
df['predicted_class'] = agglo.fit_predict(X)

df[['animal_name','class_type','predicted_class']].head()
```

4. Compute the mean squared error by comparing the actual class and predicted
high-level class.

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(df['class_type'], df['predicted_class'])

print("Mean Squared Error:", mse)
```
