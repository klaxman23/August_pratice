Case Study – 2
Objective:
• Use k-means clustering for image segmentation and to identify the dominant
color in the image.
Question:
1. Open and display the image “dog.jpeg”. Convert the image into a numpy array,
so that it can be used in further processing.
[Hint: Use the PIL module from python]
2. Find out the dimensions of the image and convert it into a two-dimensional
array.
3. Use K-means clustering with k set to 3 and cluster the image.
 [Hint: Refer to k-means module of scikit learn]
4. Predict the cluster label of every pixel in the image and plot it back as an
image.
5. Find out the three dominant colors in the image.
[Hint: The cluster centers should correspond to three dominant colors]

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load image
img = Image.open("dog.jpeg")

# Display image
plt.imshow(img)
plt.axis('off')
plt.title("Original Image")
plt.show()

# Convert to numpy array
img_array = np.array(img)
```

```python
# Image dimensions
print("Image shape:", img_array.shape)

# Reshape to 2D array
pixels = img_array.reshape(-1, 3)
print("Reshaped pixels:", pixels.shape)
```

```python
from sklearn.cluster import KMeans

# Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(pixels)
```

```python
# Predict cluster labels
labels = kmeans.predict(pixels)

# Replace pixels with cluster centers
segmented_pixels = kmeans.cluster_centers_[labels]

# Convert back to image shape
segmented_image = segmented_pixels.reshape(img_array.shape).astype(np.uint8)

# Display segmented image
plt.imshow(segmented_image)
plt.axis('off')
plt.title("Segmented Image (K=3)")
plt.show()
```

```python
# Dominant colors
dominant_colors = kmeans.cluster_centers_.astype(int)

print("Dominant Colors (RGB):")
for idx, color in enumerate(dominant_colors):
    print(f"Color {idx+1}: {color}")
```
