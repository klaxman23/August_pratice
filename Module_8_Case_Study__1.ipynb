1. Scikit learn comes with the pre-loaded dataset, load the digits dataset from
that collection and write a helper function to plot the image using matplotlib.
[Hint: Explore datasets module from scikit learn]


Step 1: Load the digits dataset from scikit-learn

```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

# Load dataset
digits = load_digits()

# Data and images
X = digits.data
y = digits.target
images = digits.images
```

Step 2: Helper function to plot a digit image

```python
def plot_digit(image, label=None):
    plt.imshow(image, cmap='gray')
    if label is not None:
        plt.title(f"Label: {label}")
    plt.axis('off')
```

Step 3: Plot a sample digit

```python
plot_digit(images[0], y[0])
plt.show()
```

2. Make a train-test split with 20% of the data set aside for testing. Fit a logistic
regression model and observe the accuracy.


Step 1: Train–Test Split (80% Train, 20% Test)

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

Step 2: Fit Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

lr = LogisticRegression(max_iter=3000)
lr.fit(X_train, y_train)
```

Step 3: Predict & Compute Accuracy

```python
y_pred = lr.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print("Logistic Regression Accuracy:", accuracy)
```

3. Using scikit learn perform a PCA transformation such that the transformed
dataset can explain 95% of the variance in the original dataset. Find out the
number of components in the projected subspace.
[Hint: Refer to decomposition module of scikit learn]


Step 1: Import PCA from scikit-learn

```python
from sklearn.decomposition import PCA
```

Step 2: Apply PCA to retain 95% variance

```python
pca = PCA(n_components=0.95, random_state=42)
pca.fit(X_train)
```

Step 3: Number of components after PCA

```python
print("Original number of features:", X_train.shape[1])
print("Number of components after PCA:", pca.n_components_)
```

Question 4: Logistic Regression After PCA & Accuracy Comparison

Step 1: Transform Train and Test Data Using PCA

```python
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
```

Step 2: Fit Logistic Regression on PCA-Transformed Data

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

lr_pca = LogisticRegression(max_iter=3000)
lr_pca.fit(X_train_pca, y_train)
```

Step 3: Predict and Compute Accuracy

```python
y_pred_pca = lr_pca.predict(X_test_pca)
accuracy_pca = accuracy_score(y_test, y_pred_pca)

print("Accuracy after PCA:", accuracy_pca)
```

5. Compute the confusion matrix and count the number of instances that have

Question 5: Confusion Matrix & Misclassified Digits Visualization

Step 1: Compute Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
import numpy as np
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred_pca)
print("Confusion Matrix:\n", cm)
```

Step 2: Count Wrong Predictions

```python
wrong_indices = np.where(y_pred_pca != y_test)[0]
print("Number of misclassified samples:", len(wrong_indices))
```

Step 3: Plot Misclassified Digits with True & Predicted Labels

```python
plt.figure(figsize=(12, 8))

for i, idx in enumerate(wrong_indices[:10]):  # display first 10 errors
    plt.subplot(2, 5, i + 1)
    plt.imshow(X_test[idx].reshape(8, 8), cmap="gray")
    plt.title(f"True: {y_test[idx]} | Pred: {y_pred_pca[idx]}")
    plt.axis("off")

plt.tight_layout()
plt.show()
```
