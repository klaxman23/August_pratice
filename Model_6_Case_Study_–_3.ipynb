Import Libraries & Load Data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
df = pd.read_csv("FyntraCustomerData.csv")

# Quick look
df.head()
```

1. Jointplot – Time on Website vs Yearly Amount Spent

```python
sns.jointplot(
    x="Time on Website",
    y="Yearly Amount Spent",
    data=df,
    kind="scatter"
)
plt.show()
```

There is a correlation, but it is weak to moderate

Points are more scattered

2. Jointplot – Time on App vs Yearly Amount Spent

```python
sns.jointplot(
    x="Time on App",
    y="Yearly Amount Spent",
    data=df,
    kind="scatter"
)
plt.show()
```

Yes, this correlation is stronger than Time on Website
Points follow a clearer upward trend
Business Insight:
Customers spending more time on the mobile app tend to spend more money.

3. Pairplot – Explore Relationships Across Dataset

```python
sns.pairplot(df)
plt.show()
```

From the pairplot:

Length of Membership shows the strongest correlation with
Yearly Amount Spent

Time on App is the second strongest
Long-term customers spend more regardless of platform.

Linear Model Plot – Length of Membership vs Yearly Amount Spent

```python
sns.lmplot(
    x="Length of Membership",
    y="Yearly Amount Spent",
    data=df
)
plt.show()
```

Yes, the data fits very well in a linear plot

Strong linear relationship is visible

5. Train–Test Split & Role of random_state=85

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Features and target
X = df.drop("Yearly Amount Spent", axis=1)
y = df["Yearly Amount Spent"]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=85
)
```

What is the use of random_state=85?

Ensures reproducibility

Same data split every time you run the code

Helpful for debugging and fair evaluation

6. Predict & Scatter Plot (Actual vs Predicted)

```python
# Train model
lm = LinearRegression()
lm.fit(X_train, y_train)

# Predictions
predictions = lm.predict(X_test)

# Scatter plot
plt.scatter(y_test, predictions)
plt.xlabel("Actual Yearly Amount Spent")
plt.ylabel("Predicted Yearly Amount Spent")
plt.title("Actual vs Predicted Values")
plt.show()
```

Points lie close to the diagonal

Indicates good prediction accuracy

Model performs well

Root Mean Squared Error (RMSE)

```python
from sklearn.metrics import mean_squared_error

rmse = np.sqrt(mean_squared_error(y_test, predictions))
rmse
```

RMSE value = printed output (typically low, indicating good model fit)

Lower RMSE → better predictive performance
