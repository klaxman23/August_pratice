1. Data Loading:
a. Load the dataset ‚Äúprisoners.csv‚Äù using pandas and display the first and
last five rows in the dataset.
[Hint: Refer to read_csv, head and tail methods in pandas]


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("prisoners.csv")

# Display first 5 rows
print("First 5 rows:")
print(df.head())

# Display last 5 rows
print("\nLast 5 rows:")
print(df.tail())
```

Describe the dataset & analyze zero-inmate rows

```python
# Statistical description
print("\nDataset Description:")
print(df.describe())

# Number of columns
print("\nNumber of columns:", df.shape[1])

# Rows where all benefit values are zero
zero_inmates = df.loc[
    (df.iloc[:, 1:] == 0).all(axis=1)
]

print("\nStates with zero inmates benefitted:")
print(zero_inmates)
```

üîπ 2. Data Manipulation
a) Create a new column total_benefitted

```python
# Create total_benefitted column
df["total_benefitted"] = df.iloc[:, 1:].sum(axis=1)

print("\nData with total_benefitted column:")
print(df.head())
```

b) Create a new row "totals" (sum across all states)

```python
# Create totals row
totals = df.iloc[:, 1:].sum()
totals["STATE/UT"] = "Total"

# Append totals row
df = pd.concat([df, pd.DataFrame([totals])], ignore_index=True)

print("\nDataset after adding totals row:")
print(df.tail())
```

a) Bar Plot ‚Äì Total Benefitted Inmates per State

```python
# Remove totals row for plotting
plot_df = df[df["STATE/UT"] != "Total"]

plt.figure(figsize=(14, 6))
plt.bar(plot_df["STATE/UT"], plot_df["total_benefitted"])
plt.xticks(rotation=90)
plt.xlabel("State / UT")
plt.ylabel("Total Benefitted Inmates")
plt.title("Total Benefitted Prisoners by State")
plt.tight_layout()
plt.show()

# State with maximum beneficiaries
max_state = plot_df.loc[
    plot_df["total_benefitted"].idxmax(), "STATE/UT"
]

print("State with maximum beneficiaries:", max_state)
```

b) Pie Chart ‚Äì Ratio of Different Benefit Modes

```python
# Sum of each benefit mode (excluding STATE/UT and total_benefitted)
benefit_modes = df[df["STATE/UT"] != "Total"].iloc[:, 1:-1].sum()

plt.figure(figsize=(8, 8))
plt.pie(
    benefit_modes,
    labels=benefit_modes.index,
    autopct="%1.1f%%",
    startangle=140
)
plt.title("Distribution of Prisoner Benefits by Mode")
plt.show()
```
