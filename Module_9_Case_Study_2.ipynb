Objective:
• Practice Naive Bayes algorithm-based classification.
• Identify the predictors that can be of influence by experiment.
Questions:
1. Load the kinematics dataset as measured on mobile sensors from the file
“run_or_walk.csv”. List out the columns in the dataset.
2. Let the target variable ‘y’ be the activity and assign all the columns after it to
‘x’.
3. Using Scikit-learn fit a Gaussian Naive Bayes model and observe the accuracy.
Generate a classification report using scikit learn.
4. Repeat the model once using only the acceleration values as predictors and
then using only the gyro values as predictors. Comment on the difference in
accuracy between both the models.

Load the dataset and list columns

```python
import pandas as pd

# Load dataset
df = pd.read_csv("run_or_walk.csv")

# Display first few rows
df.head()
```

```python
print("Columns in dataset:")
print(df.columns.tolist())
```

Define target (y) and predictors (X)

```python
y = df['activity']

# Predictor variables (all columns after activity)
X = df.loc[:, df.columns[df.columns.get_loc('activity') + 1:]]
```

3. Gaussian Naive Bayes using all predictors

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)

# Train model
gnb = GaussianNB()
gnb.fit(X_train, y_train)

# Predictions
y_pred = gnb.predict(X_test)

# Accuracy
print("Accuracy (All Features):", accuracy_score(y_test, y_pred))

# Classification Report
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

4.Model using only Acceleration features

```python
# Acceleration features only
acc_features = ['acceleration_x', 'acceleration_y', 'acceleration_z']
X_acc = df[acc_features]

X_train_acc, X_test_acc, y_train_acc, y_test_acc = train_test_split(
    X_acc, y, test_size=0.25, random_state=42
)

# Train model
gnb_acc = GaussianNB()
gnb_acc.fit(X_train_acc, y_train_acc)

# Predictions
y_pred_acc = gnb_acc.predict(X_test_acc)

# Accuracy
print("Accuracy (Acceleration Only):", accuracy_score(y_test_acc, y_pred_acc))

# Classification Report
print("\nClassification Report (Acceleration Only):\n")
print(classification_report(y_test_acc, y_pred_acc))
```

5. Model using only Gyroscope features

```python
# Gyroscope features only
gyro_features = ['gyro_x', 'gyro_y', 'gyro_z']
X_gyro = df[gyro_features]

X_train_gyro, X_test_gyro, y_train_gyro, y_test_gyro = train_test_split(
    X_gyro, y, test_size=0.25, random_state=42
)

# Train model
gnb_gyro = GaussianNB()
gnb_gyro.fit(X_train_gyro, y_train_gyro)

# Predictions
y_pred_gyro = gnb_gyro.predict(X_test_gyro)

# Accuracy
print("Accuracy (Gyro Only):", accuracy_score(y_test_gyro, y_pred_gyro))

# Classification Report
print("\nClassification Report (Gyro Only):\n")
print(classification_report(y_test_gyro, y_pred_gyro))
```
