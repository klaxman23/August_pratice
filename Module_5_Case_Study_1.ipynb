4. Is there any correspondence between the critics’ evaluation of a movie and its
acceptance by the public? Find out, by plotting the net profitability of a movie
against the ratings it receives on Rotten Tomatoes.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create dataset with required columns
data = {
    "Movie": [
        "The Great Journey",
        "Space Odyssey",
        "Hidden Truth",
        "Lost Kingdom",
        "Final Mission",
        "Love in Paris",
        "Dark Streets",
        "Ocean Quest"
    ],
    "Budget": [170, 200, 80, 210, 185, 45, 65, 220],
    "Revenue": [480, 600, 140, 650, 520, 120, 150, 720],
    "RottenTomatoes": [82, 88, 71, 90, 85, 65, 70, 94]
}

df = pd.DataFrame(data)

# Calculate Net Profit
df["NetProfit"] = df["Revenue"] - df["Budget"]

# Plot Net Profitability vs Rotten Tomatoes Rating
plt.figure(figsize=(6, 5))
plt.scatter(df["RottenTomatoes"], df["NetProfit"], color="blue")

plt.xlabel("Rotten Tomatoes Rating (%)")
plt.ylabel("Net Profit (Million USD)")
plt.title("Net Profitability vs Critics Rating")

plt.grid(True)
plt.show()
```

5.1: From the raw data below create a data frame
'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
'last_name': ['Miller', 'Jacobson', ".", 'Milner', 'Cooze'],
'age': [42, 52, 36, 24, 73],
'preTestScore': [4, 24, 31, ".", "."],
'postTestScore': ["25,000", "94,000", 57, 62, 70]


```python
import pandas as pd
import numpy as np

# Raw data
data = {
    'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
    'last_name': ['Miller', 'Jacobson', '.', 'Milner', 'Cooze'],
    'age': [42, 52, 36, 24, 73],
    'preTestScore': [4, 24, 31, '.', '.'],
    'postTestScore': ['25,000', '94,000', 57, 62, 70]
}

# Create DataFrame
df = pd.DataFrame(data)

# Replace "." with NaN
df.replace('.', np.nan, inplace=True)

# Remove commas and convert postTestScore to numeric
df['postTestScore'] = (
    df['postTestScore']
    .astype(str)
    .str.replace(',', '')
    .astype(float)
)

# Convert preTestScore to numeric
df['preTestScore'] = pd.to_numeric(df['preTestScore'])

print(df)
```

6.3: - Create a series from the raw data below
'India_is_big', 'Population_is_huge', np.nan, 'Has_diverse_culture'
a) split the individual strings wherever ‘_’ comes and create a list out of it.
b) Access the individual elements of a list
c) Expand the elements so that all individual elements get split by ‘_’ and instead
of list returns individual elements


```python
import pandas as pd
import numpy as np

# Raw data
data = ['India_is_big', 'Population_is_huge', np.nan, 'Has_diverse_culture']

# Create Pandas Series
s = pd.Series(data)

print("Original Series:")
print(s)

# ------------------------------------------------
# a) Split strings wherever '_' comes → list
# ------------------------------------------------
split_series = s.str.split('_')

print("\nAfter splitting on '_' (list output):")
print(split_series)

# ------------------------------------------------
# b) Access individual elements of the list
# (example: first element of each split)
# ------------------------------------------------
print("\nAccess first element from each list:")
print(split_series.str[0])

# ------------------------------------------------
# c) Expand elements so each split becomes a column
# ------------------------------------------------
expanded_df = s.str.split('_', expand=True)

print("\nExpanded Data (each split as separate column):")
print(expanded_df)
```

6.4: Create a series and replace either X or dog with XX-XX
'A', 'B', 'C', 'AabX', 'BacX','', np.nan, 'CABA', 'dog', 'cat'

```python
import pandas as pd
import numpy as np

# Raw data
data = ['A', 'B', 'C', 'AabX', 'BacX', '', np.nan, 'CABA', 'dog', 'cat']

# Create Pandas Series
s = pd.Series(data)

print("Original Series:")
print(s)

# Replace 'X' or 'dog' with 'XX-XX'
result = s.replace(to_replace=r'X|dog', value='XX-XX', regex=True)

print("\nAfter replacing 'X' or 'dog' with 'XX-XX':")
print(result)
```

6.6:- Create a series and reverse all lower case words
'India 1998', 'big country', np.nan

```python
import pandas as pd
import numpy as np
import re

# Raw data
data = ['India 1998', 'big country', np.nan]

# Create Pandas Series
s = pd.Series(data)

print("Original Series:")
print(s)

# Function to reverse lowercase words
def reverse_lowercase(text):
    if pd.isna(text):
        return text
    return re.sub(r'\b[a-z]+\b', lambda x: x.group()[::-1], text)

# Apply function
result = s.apply(reverse_lowercase)

print("\nAfter reversing lowercase words:")
print(result)
```

6.9: Create pandas series and print in three columns value 0 or 1 is a or b or c exists
in values
'a', 'a|b', np.nan, 'a|c'


```python
import pandas as pd
import numpy as np

# Raw data
data = ['a', 'a|b', np.nan, 'a|c']

# Create Pandas Series
s = pd.Series(data)

print("Original Series:")
print(s)

# Create DataFrame with presence of a, b, c
result = pd.DataFrame({
    'a': s.str.contains('a', na=False).astype(int),
    'b': s.str.contains('b', na=False).astype(int),
    'c': s.str.contains('c', na=False).astype(int)
})

print("\nPresence of a, b, c (1 = exists, 0 = does not exist):")
print(result)
```

6.10: Create pandas dataframe having keys and ltable and rtable as below -
'key': ['One', 'Two'], 'ltable': [1, 2]
'key': ['One', 'Two'], 'rtable': [4, 5]
Merge both the tables based on key

```python
import pandas as pd

# Create left table
left_df = pd.DataFrame({
    'key': ['One', 'Two'],
    'ltable': [1, 2]
})

# Create right table
right_df = pd.DataFrame({
    'key': ['One', 'Two'],
    'rtable': [4, 5]
})

# Merge both tables based on key
merged_df = pd.merge(left_df, right_df, on='key')

# Print merged DataFrame
print(merged_df)
```
