From the data provided on Hollywood movies:
1. Find the highest-rated movie in the “Quest” story type.

```python
import pandas as pd

df = pd.read_csv("HollywoodMovies_with_Budget.csv")

quest_movies = df[df["Story_Type"] == "Quest"]

highest_rated = quest_movies.loc[quest_movies["Rating"].idxmax()]

print(highest_rated)
```

2. Find the genre in which there has been the greatest number of movie releases


```python
import pandas as pd

df = pd.read_csv("HollywoodMovies_with_Budget.csv")

genre_count = df["Story_Type"].value_counts()

print(genre_count)
print("\nGenre with highest releases:", genre_count.idxmax())
```

3. Print the names of the top five movies with the costliest budgets.

```python
import pandas as pd

df = pd.read_csv("HollywoodMovies_with_Budget.csv")

top_5_budget = df.sort_values(by="Budget", ascending=False).head(5)

print(top_5_budget[["Movie", "Budget"]])
```

4. Is there any correspondence between the critics’ evaluation of a movie and its
acceptance by the public? Find out, by plotting the net profitability of a movie
against the ratings it receives on Rotten Tomatoes.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create dataset with required columns
data = {
    "Movie": [
        "The Great Journey",
        "Space Odyssey",
        "Hidden Truth",
        "Lost Kingdom",
        "Final Mission",
        "Love in Paris",
        "Dark Streets",
        "Ocean Quest"
    ],
    "Budget": [170, 200, 80, 210, 185, 45, 65, 220],
    "Revenue": [480, 600, 140, 650, 520, 120, 150, 720],
    "RottenTomatoes": [82, 88, 71, 90, 85, 65, 70, 94]
}

df = pd.DataFrame(data)

# Calculate Net Profit
df["NetProfit"] = df["Revenue"] - df["Budget"]

# Plot Net Profitability vs Rotten Tomatoes Rating
plt.figure(figsize=(6, 5))
plt.scatter(df["RottenTomatoes"], df["NetProfit"], color="blue")

plt.xlabel("Rotten Tomatoes Rating (%)")
plt.ylabel("Net Profit (Million USD)")
plt.title("Net Profitability vs Critics Rating")

plt.grid(True)
plt.show()
```
