```python
import pandas as pd
import matplotlib.pyplot as plt

# ============================================================
# STEP 1: Load BigMart Sales Data
# ============================================================

# Make sure BigMartSalesData.csv is in the same folder
df = pd.read_csv("BigMartSalesData.csv")

# Convert InvoiceDate to datetime
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])

# Filter data for year 2011
df_2011 = df[df["InvoiceDate"].dt.year == 2011]

# Create Month column
df_2011["Month"] = df_2011["InvoiceDate"].dt.month


# ============================================================
# QUESTION 1
# Total Sales Per Month (Line Plot)
# ============================================================

monthly_sales = df_2011.groupby("Month")["TotalAmount"].sum()

plt.figure()
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.title("Total Sales Per Month - 2011")
plt.show()

print("Lowest sales month:", monthly_sales.idxmin())


# ============================================================
# QUESTION 2
# Total Sales Per Month (Bar Chart)
# ============================================================

plt.figure()
bars = plt.bar(monthly_sales.index, monthly_sales.values)
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.title("Total Sales Per Month - 2011 (Bar Chart)")

# Enhancement 1: Show value on bars
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f"{bar.get_height():.0f}",
        ha="center",
        va="bottom",
        fontsize=8
    )

plt.show()


# ============================================================
# QUESTION 3
# Country-wise Sales (Pie Chart)
# ============================================================

country_sales = df_2011.groupby("Country")["TotalAmount"].sum()

plt.figure()
plt.pie(
    country_sales.values,
    labels=country_sales.index,
    autopct="%1.1f%%",
    shadow=True,          # Enhancement
    startangle=90         # Enhancement
)
plt.title("Country-wise Sales Contribution - 2011")
plt.show()

print("Highest contributing country:", country_sales.idxmax())


# ============================================================
# QUESTION 4
# Scatter Plot - Invoice Amount Concentration
# ============================================================

plt.figure()
plt.scatter(
    df_2011["InvoiceNo"],
    df_2011["TotalAmount"],
    color="green",        # Enhancement
    alpha=0.5
)
plt.xlabel("Invoice Number")
plt.ylabel("Invoice Amount")
plt.title("Invoice Amount Distribution - 2011")
plt.show()

print(
    "Most invoice amounts are concentrated in range:",
    df_2011["TotalAmount"].quantile([0.25, 0.75])
)
```
