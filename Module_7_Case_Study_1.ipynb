Case Study â€“ 1: Voice Gender Classification (Logistic Regression)

```python
# Import libraries
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# Load dataset
df = pd.read_csv("voice.csv")

# Display first few rows
print(df.head())

# Identify target variable
X = df.drop("label", axis=1)
y = df["label"]

# Encode target variable (male/female -> 0/1)
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Train-test split (20% test data)
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42
)

print("Training samples:", X_train.shape)
print("Testing samples:", X_test.shape)
```

Fit Logistic Regression model and evaluate accuracy

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Create Logistic Regression model
log_reg = LogisticRegression(max_iter=3000)

# Train model
log_reg.fit(X_train, y_train)

# Predictions
y_pred = log_reg.predict(X_test)

# Accuracy
accuracy_full = accuracy_score(y_test, y_pred)
print("Accuracy with all features:", accuracy_full)
```

Compute correlation matrix & visualize using seaborn heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute correlation matrix
corr_matrix = X.corr()

# Plot heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix of Voice Features")
plt.show()
```

Remove highly correlated predictors and refit model

```python
# Set correlation threshold
threshold = 0.90

# Find correlated feature pairs
corr_features = set()

for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            colname = corr_matrix.columns[i]
            corr_features.add(colname)

print("Highly correlated features removed:", corr_features)

# Drop correlated features
X_reduced = X.drop(columns=corr_features)

# Train-test split again
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(
    X_reduced, y_encoded, test_size=0.2, random_state=42
)

# Train Logistic Regression again
log_reg_reduced = LogisticRegression(max_iter=3000)
log_reg_reduced.fit(X_train_r, y_train_r)

# Predictions
y_pred_r = log_reg_reduced.predict(X_test_r)

# Accuracy after feature reduction
accuracy_reduced = accuracy_score(y_test_r, y_pred_r)

print("Accuracy after removing correlated features:", accuracy_reduced)
```
