```python
import pandas as pd

df = pd.read_csv("SFO_Salaries_2011_2014.csv")

increase = (
    df[df["Year"] == 2014]["TotalPayBenefits"].sum()
    - df[df["Year"] == 2011]["TotalPayBenefits"].sum()
)

print("Total salary cost increase from 2011 to 2014:", increase)
```

Which Job Title in the Year 2014 has the highest mean salary?

```python
result = (
    df[df["Year"] == 2014]
    .groupby("JobTitle")["TotalPayBenefits"]
    .mean()
)

print(result)
print("\nHighest Mean Salary Job:", result.idxmax())
```

3.How much money could have been saved in the Year 2014 by stopping OverTimePay?

```python
savings = df[df["Year"] == 2014]["OvertimePay"].sum()

print("Money saved by stopping OverTimePay in 2014:", savings)
```

4. Which are the top 5 common jobs in the Year 2014 and how much do they cost SFO?

```python
top_jobs_cost = (
    df[df["Year"] == 2014]
    .groupby("JobTitle")["TotalPayBenefits"]
    .sum()
    .sort_values(ascending=False)
    .head(5)
)

print(top_jobs_cost)
```

5. Who was the top earning employee across all the years? Enhancements for code You can try these enhancements in code

```python
top_employee = (
    df.groupby("EmployeeName")["TotalPayBenefits"]
    .sum()
    .idxmax()
)

top_salary = (
    df.groupby("EmployeeName")["TotalPayBenefits"]
    .sum()
    .max()
)

print("Top earning employee:", top_employee)
print("Total earnings:", top_salary)
```

6. Which are the last 5 common jobs in the Year 2014 and how much do they cost SFO?

```python
last_jobs_cost = (
    df[df["Year"] == 2014]
    .groupby("JobTitle")["TotalPayBenefits"]
    .sum()
    .sort_values()
    .head(5)
)

print(last_jobs_cost)
```

8. Which Job Title in the Year 2014 has the lowest mean salary?

```python
lowest_mean_job = (
    df[df["Year"] == 2014]
    .groupby("JobTitle")["TotalPayBenefits"]
    .mean()
)

print(lowest_mean_job)
print("\nLowest mean salary job (2014):", lowest_mean_job.idxmin())
```
